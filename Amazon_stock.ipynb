```python
import yfinance as yahooFinance
import matplotlib.pyplot as plt

def plot_sprice(ticker, years=10):

    stock_data = yahooFinance.download(ticker, period=f"{years}y")


    close_prices = stock_data['Close']


    plt.figure(figsize=(10, 6))
    close_prices.plot()
    plt.title(f"{ticker} Stock Price Over the Last Decade")
    plt.xlabel("Year")
    plt.ylabel("Stock Price ($USD)")
    plt.grid(True)
    plt.show()


ticker_symbol = 'AMZN'


plot_sprice(ticker_symbol, years=10)
```

```python

import yfinance as yahooFinance
import matplotlib.pyplot as plt

def plot_msprices(tickers, start_date, end_date, shaded_start, shaded_end):
    plt.figure(figsize=(10, 6))

    for ticker in tickers:
        stock_data = yahooFinance.download(ticker, start=start_date, end=end_date)
        close_prices = stock_data['Close']
        net_change = close_prices.iloc[-1] - close_prices.iloc[0]
        close_prices.plot(label=f'{ticker} (Stock Price Net Change: ${net_change:.2f})')

    # Plot shaded region for the specified date range (shade stands for COVID19 pandemic period before the vaccine was realeased in December 2020)
    plt.axvspan(shaded_start, shaded_end, color='gray', alpha=0.3)

    plt.title("Stock Prices Comparison During COVID19 Pandemic Pre-Vaccine Release (2020-02-23 - 2020-12-11)")
    plt.xlabel("Date")
    plt.ylabel("Stock Price ($USD)")
    plt.grid(True)
    plt.legend()
    plt.show()

tickers = ['TSLA', 'SONY', 'AMZN']
start_date = '2018-02-23'
end_date = '2021-03-20'
shaded_start = '2020-02-23'
shaded_end = '2021-01-01'

plot_msprices(tickers, start_date, end_date, shaded_start, shaded_end)
```

```python
import yfinance as yahooFinance
import matplotlib.pyplot as plt

def plot_msprices(tickers, start_date, end_date, shaded_start, shaded_end):
    plt.figure(figsize=(10, 6))

    for ticker in tickers:
        stock_data = yahooFinance.download(ticker, start=start_date, end=end_date)
        close_prices = stock_data['Close']

        # Calculate net percentage change
        net_change_percentage = ((close_prices.iloc[-1] - close_prices.iloc[0]) / close_prices.iloc[0]) * 100

        close_prices.plot(label=f'{ticker} (Net % Change: {net_change_percentage:.2f}%)')

    # Plot shaded region for the specified date range
    plt.axvspan(shaded_start, shaded_end, color='gray', alpha=0.3)

    plt.title("Stock Prices Comparison During 2022 Bear Market (2022-01-01 - 2022-10-31)")
    plt.xlabel("Date")
    plt.ylabel("Stock Price ($USD)")
    plt.grid(True)
    plt.legend()
    plt.show()

tickers = ['TSLA', 'SONY', 'AMZN']
start_date = '2021-12-01'
end_date = '2023-01-01'
shaded_start = '2022-01-01'
shaded_end = '2022-10-31'

plot_msprices(tickers, start_date, end_date, shaded_start, shaded_end)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def plot_sprice_with_max(ticker, years=10):
    stock_data = yf.download(ticker, period=f"{years}y")
    close_prices = stock_data['Close']

    max_price = close_prices.max()  # Find the maximum price
    max_date = close_prices.idxmax()  # Find the date of the maximum price

    plt.figure(figsize=(10, 6))
    close_prices.plot()
    plt.title(f"{ticker} Stock Price Over the Last Decade")
    plt.xlabel("Date")
    plt.ylabel("Stock Price ($USD)")
    plt.grid(True)

    # Annotate the maximum price
    plt.annotate(f'Max Price: ${max_price:.2f}\nDate: {max_date.strftime("%Y-%m-%d")}',
                 xy=(max_date, max_price),
                 xytext=(max_date, max_price*0.9),
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 )

    plt.show()

ticker_symbol = 'AMZN'
plot_sprice_with_max(ticker_symbol)
```

```python
import yfinance as yahooFinance
import matplotlib.pyplot as plt

def plot_msprices(tickers, start_date, end_date, shaded_start, shaded_end):
    plt.figure(figsize=(10, 6))

    for ticker in tickers:
        stock_data = yahooFinance.download(ticker, start=start_date, end=end_date)
        close_prices = stock_data['Close']
        net_changes = close_prices.diff().dropna()  # Calculate net change after each closing time
        net_changes.plot(label=f'{ticker} (Net Change)')

    # Plot shaded region for the specified date range
    plt.axvspan(shaded_start, shaded_end, color='gray', alpha=0.3)

    plt.title("Net Change in Stock Prices During 2022 Bear Market (2022-01-01 - 2022-10-31)")
    plt.xlabel("Date")
    plt.ylabel("Net Change ($USD)")
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
tickers = ['TSLA', 'SONY', 'AMZN']
start_date = '2021-12-01'
end_date = '2023-01-01'
shaded_start = '2022-01-01'
shaded_end = '2022-10-31'

plot_msprices(tickers, start_date, end_date, shaded_start, shaded_end)
```
